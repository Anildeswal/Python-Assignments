```python
#Ques1: Explain GET and POST methods.
'''GET and POST are two of the most commonly used HTTP request methods.

GET requests are used to retrieve data from a server. When a client (such as a web browser) makes a GET request, it sends a request to the server for a specific 
resource (such as a web page or an image), and the server responds with the requested data. GET requests are typically used for simple queries where no changes 
are made to the server, such as searching or viewing a page.

POST requests, on the other hand, are used to submit data to a server. When a client makes a POST request, it sends data to the server to be processed (such as 
form data or user input), and the server responds with a confirmation of the success or failure of the request. POST requests are typically used for more complex 
operations, such as submitting a form or updating data on the server.

The main difference between the two methods is that GET requests only retrieve data, while POST requests can modify data on the server. Additionally, GET requests
are generally considered less secure than POST requests, as data is sent in the URL, which can be seen by anyone who has access to the client or server logs. POST
requests, on the other hand, send data in the request body, which is not visible in the URL.'''
```

```python
#Ques2: Why is request used in Flask?
'''Requests are used in Flask to handle HTTP requests made by clients (such as web browsers) to the Flask application. In Flask, a request is an object that 
contains information about the incoming HTTP request, such as the HTTP method (GET, POST, etc.), the headers, and any data that was included in the request body.
The Flask application can then use this information to generate a response to send back to the client. Using the request object in Flask, you can access various
properties of the incoming request, such as request method, headers, URL parameters, and request body. You can use this information to perform different actions,
such as processing form data, validating user input, and generating a response to send back to the client.

Overall, the request object in Flask is a very useful tool for handling HTTP requests in your Flask application, and it makes it easy to process and respond to 
client requests in a flexible and efficient way.'''
```

```python
#Ques3: Why is redirect() used in Flask?
'''The redirect() function is used in Flask to redirect a client (such as a web browser) to a different URL. In many cases, you may want to redirect a user to a different page after they perform some action on your website, such as submitting a form or logging in. 
Instead of simply rendering a new template with the new page content, you can use the redirect() function to send the user to a completely different URL. This can
be useful for maintaining a clean URL structure and preventing issues with form resubmissions. The redirect() function is a useful tool in Flask for redirecting 
users to different pages and managing the flow of your application.'''

from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return redirect(url_for('new_page'))

@app.route('/new_page')
def new_page():
    return 'This is the new page!'

if __name__ == '__main__':
    app.run(host=0.0.0.0)
```

```python
#Ques5: Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

from flask import Flask,request ,render_template , jsonify

app = Flask(__name__)

@app.route('/')
def home_page():
    return render_template('index.html')

@app.route('/postman_action',methods=['POST'])
def math_ops1():
    if(request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
            
        return jsonify(result)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
