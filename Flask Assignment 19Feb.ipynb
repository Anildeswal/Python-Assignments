```python
#Ques2: Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello World!!</h1>"
if __name__=="__main__":
    app.run(host="0.0.0.0",port=8080)
```

```python
#Ques3: What is App routing in Flask? Why do we use app routes?
'''In Flask, app routing refers to the process of mapping URLs to view functions. When a user requests a specific URL, Flask uses the app routing system to 
determine which view function should handle the request. To define an app route in Flask, you use the @app.route decorator. This decorator maps a URL to a Python function that should be called when that URL is requested.
app routing is a key feature of Flask that allows you to define the structure and behavior of your web application by mapping URLs to view functions. It makes it
easy to create a logical structure for your application and handle different types of HTTP requests.App routing is important in Flask because it allows you to 
define the structure and behavior of your web application. By mapping URLs to view functions, you can create a logical structure for your application and make it 
easy for users to navigate.App routes also allow you to handle different types of HTTP requests, such as GET, POST, PUT, and DELETE. By using different decorators,
you can define separate view functions to handle each type of request. This makes it easy to create RESTful web services or to implement CRUD operations in your 
application.'''
```

```python
#Ques4: Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:-
# Company Name: ABC Corporation
# Location: India
# Contact Detail: 999-999-9999

from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return '<h1>Welcome to ABC Corporation </h1>'

@app.route('/')
def index():
    return "Company Name: ABC Corporation  \n  Location: India  \n  Contact Detail: 999-999-9999"

if __name__ == '__main__':
    app.run(host="0.0.0.0",port=8080)
```

```python
#Ques5: What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
'''In Flask, the url_for() function is used to build URLs for specific routes based on their names. This function takes the name of the view function and any 
arguments that the function expects and returns a URL that can be used to navigate to that route.'''

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/user/<username>')
def show_user_profile(username):
    return 'User %s' % username

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return 'Post %d' % post_id

with app.test_request_context():
    print(url_for('show_user_profile', username='Anil deswal'))
    print(url_for('show_post', post_id=123))
```
